```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data
death_tolls = pd.read_csv("us_monthly_newdeaths.csv")

# Create a figure and subplots
fig, axes = plt.subplots(4, 1, figsize=(6, 8), sharex=True)
fig.subplots_adjust(hspace=0.5)

# Define colors
bar_color = '#f21462'
highlight_color = '#ff1a1a'

# Create bar plots for each quarter
for i, ax in enumerate(axes):
    start_index = i * 3
    end_index = start_index + 3
    quarter_data = death_tolls[start_index:end_index]
    ax.bar(quarter_data['Month'], quarter_data['New_deaths'], color=bar_color, alpha=0.5)
    ax.set_yticklabels([])
    ax.tick_params(bottom=0)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['left'].set_visible(False)
    
    # Highlight specific quarters
    ax.bar(quarter_data['Month'][1], quarter_data['New_deaths'][1], color=highlight_color, linewidth=2.6)
    
    # Add annotations
    ax.text(quarter_data['Month'][0], quarter_data['New_deaths'][0] + 50, str(int(quarter_data['New_deaths'][0])), ha='center', color=bar_color)
    ax.text(quarter_data['Month'][1], quarter_data['New_deaths'][1] + 50, str(int(quarter_data['New_deaths'][1])), ha='center', color=highlight_color, weight='bold')
    
    # Highlight cumulative deaths
    ax.text(quarter_data['Month'][2], quarter_data['New_deaths'][2] + 50, "Cumulative deaths: " + str(int(death_tolls.iloc[end_index-1]['Cumulative Deaths'])), ha='center', color=bar_color)

# Set labels and title
axes[-1].set_xticklabels(death_tolls['Month'], rotation=45, ha='right')
axes[-1].set_xlabel('Month')
fig.suptitle('Monthly New Deaths in the US', fontsize=16, weight='bold')

# Display the plot
plt.show()
```
