# Story Telling

## File Name : "month_sales.csv"

### Demo Output :

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the new dataset
monthly_sales = pd.read_csv("month_sales.csv")  
monthly_sales
```

```python
# Create subplots
fig, axes = plt.subplots(4, 1, figsize=(6, 8))

# Plot data for each subplot
for ax in axes:
    ax.plot(monthly_sales['Months'], monthly_sales['sales'], color='#17a589', alpha=0.5)  # Updated line color
    ax.set_xticklabels([])
    ax.set_yticklabels([])
    ax.grid(False)
    ax.tick_params(bottom=0, left=0)
    for spine in ['left', 'right', 'top', 'bottom']:
        ax.spines[spine].set_visible(False)
        
# Highlight specific quarters
axes[0].plot(monthly_sales['Months'][0:3], monthly_sales['sales'][0:3], color='#c0392b', linewidth=2.6)  
axes[1].plot(monthly_sales['Months'][2:6], monthly_sales['sales'][2:6], color='#c0392b', linewidth=2.6)  
axes[2].plot(monthly_sales['Months'][5:9], monthly_sales['sales'][5:9], color='#c0392b', linewidth=2.6)  
axes[3].plot(monthly_sales['Months'][8:12], monthly_sales['sales'][8:12], color='#c0392b', linewidth=2.6)  

axes[0].text(1, 5000, '0', alpha=0.5)
axes[0].text(2.5, 50000, '77', alpha=0.5)
axes[1].text(2.5, 50000, '77', alpha=0.5)
axes[1].text(5.9, 25000, '785', alpha=0.5)
axes[2].text(5.9, 25000, '785', alpha=0.5)
axes[2].text(8.8, 85000, '739', alpha=0.5)
axes[3].text(8.8, 85000, '739', alpha=0.5)
axes[3].text(12, 75000, '2247', alpha=0.5)

# Month labels
axes[0].text(2.2, 5000, 'Jan-Mar', rotation=60, weight="bold", color="Red")
axes[1].text(3.7, 17000, 'Apr-Jun', rotation=-10, weight="bold", color="Red")
axes[2].text(7, 21000, 'Jul-Sep', rotation=0, weight="bold", color="Red")
axes[3].text(9.7, 45000, 'Oct-Dec', rotation=50, weight="bold", color="Red")

# Header
axes[0].text(0.5, 100000, 'Monthly Sales Data', size=14, weight='bold')
axes[0].text(0.5, 90000, "Quarterly Sales Performance", size=11)

# Calculate cumulative sales
monthly_sales['Cumulative Sales'] = monthly_sales['sales'].cumsum()

# Get cumulative sales for specific quarters
cumulative_sales = list(monthly_sales.iloc[[2, 5, 8, 11], 2])

# Display cumulative sales for each subplot
for i, ax in enumerate(axes):
    ax.text(5, 1800, 'Cumulative sales = ' + str(cumulative_sales[i]))

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
