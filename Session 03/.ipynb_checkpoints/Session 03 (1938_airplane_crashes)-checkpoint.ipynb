# Example :



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the dataset
data = pd.read_csv("1938_airplane_crashes.csv", header=1)
```

```python
# Create subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 12))
```

```python
# Create subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 12))

# Calculate the total number of bars and the number of bars per month
total_bars = len(data['Months'])
bars_per_month = 1

# Adjust the width of the bars and the padding between the bars
bar_width = 0.6

# Initialize the x positions for bars
x_positions = [i * bar_width for i in range(total_bars)]

# Create a color palette for different locations
color_palette = sns.color_palette("Set3", len(data['Location'].unique()))

# Plot the data for the first subplot (airplane crash analysis)
bars = []
for i, month in enumerate(data['Months']):
    location_data = data[data['Months'] == month]
    light_color = color_palette[i % len(data['Location'].unique())]
    dark_color = sns.desaturate(light_color, 0.6)

    bar = ax1.bar(x_positions[i], location_data['Aboard'].values[0], width=bar_width / 2,
                  color=light_color, label='Aboard ' + month + ' ' + location_data['Location'].values[0])
    bars.append(bar)
    bar = ax1.bar(x_positions[i] + bar_width / 2, location_data['Fatalities'].values[0], width=bar_width / 2,
                  color=dark_color, label='Fatalities ' + month + ' ' + location_data['Location'].values[0])
    bars.append(bar)

# Annotate bars with values
for bar_group in bars:
    for rect in bar_group:
        height = rect.get_height()
        ax1.annotate(f'{height}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=9)

# Set labels and title for the first subplot
ax1.set_ylabel('Number of People')
ax1.set_title('Airplane Crash Analysis of 1938')
ax1.set_frame_on(False)
ax1.tick_params(left=False, bottom=False)
ax1.legend(loc='upper left', bbox_to_anchor=(1, 1))  # Adjust the legend position
ax1.set_xticks([])
ax1.set_yticks([])
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 12))
# Calculate cumulative deaths for each quarter
cumulative_deaths = [data['Fatalities'][:3].sum(), data['Fatalities'][3:6].sum(),
                     data['Fatalities'][6:9].sum(), data['Fatalities'][9:].sum()]

quarters = ['Jan-Mar', 'Apr-Jun', 'Jul-Sep', 'Oct-Dec']

# Plot the data for the second subplot (cumulative deaths by quarter)
bars = ax2.bar(quarters, cumulative_deaths, color=color_palette[:4])

# Annotate bars with counts
for rect in bars:
    height = rect.get_height()
    ax2.annotate(f'{int(height)}',
                 xy=(rect.get_x() + rect.get_width() / 2, height),
                 xytext=(0, 3),  # 3 points vertical offset
                 textcoords="offset points",
                 ha='center', va='bottom', fontsize=9)

ax2.set_ylabel('Cumulative Deaths')
ax2.set_title('Cumulative Deaths by Quarter')

# Remove borders and ticks from the second subplot
ax2.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)
ax2.spines['bottom'].set_visible(False)
ax2.spines['left'].set_visible(False)
ax2.tick_params(left=False, bottom=False)
ax2.set_xticks([])  # Remove x-axis ticks and labels
ax2.set_yticks([])
# Add legends of quarter months to the second subplot
legend_entries = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=color_palette[i],
                            markersize=10, label=quarter) for i, quarter in enumerate(quarters)]
ax2.legend(handles=legend_entries, loc='upper right')

plt.tight_layout()
plt.show()
```
